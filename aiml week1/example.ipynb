## Challenge: create a 5×5 matrix, compute its transpose, multiply it by itself, find the max value in each row — using only NumPy, no loops.

```python
import numpy as np
```

```python
array = np.random.random((5,5))
print(array)
```

```python
print(array.T)
```

```python
print(array*array)
```

```python
print(array.max(axis = 1))
```
