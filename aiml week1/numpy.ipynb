### Broadcasting

```python
array = np.array([[10,20,30,40] , [50,60,70,80]])
x = 10
print(array + x)
arrayy = np.array([100,200,300,400])
print(array + arrayy)
```

### Broadcasting in conditional operations

```python
age = np.array([12,19,35,18,22,41,17,8,25,2,90])
status = np.array(["Adult" , "Minor"])
age_check = np.where(age>=18 , status[0] , status[1] )
print(age_check)
```

### Broadcasting for array manipulation

```python
m = np.array([[1, 2], [3, 4]]).reshape(2,2)
v = np.array([10, 20]).reshape(1,2)
res = m * v
print(res)
```

m*v and v*m is used according to the demensions to meet the compatibility rule
