```python
def read_csv(file_name):
    with open(file_name , "r") as files:
        lines = files.readlines() 
        
        #column names
        header = lines[0].strip().split(",")
        print("Header's Name:")
        print(header)

        #5 rows
        print("\n5 rows:")
        for line in lines[1:6]:
            print(line.strip().split(","))

        #no of rows
        print("\nTotal no. of rows:", len(lines)-1)


read_csv("quikr_car.csv")
        

        
```

```python
class DataLoader:
    def __init__(self,filename):
        self.filename = filename 
        self.rows = []
        self.load_file()

    def load_file(self):
        with open(self.filename,"r") as file:
            lines = file.readlines()

        header = lines[0].strip().split(",")

        for line in lines[1: ]:
            values = line.strip().split(",")

            row_dict = {}

            n = len(header)

            for i in range(n):
                row_dict[header[i]] = values[i]

            self.rows.append(row_dict)

    def display_rows(self):
        for row in self.rows:
            print(row)

data = DataLoader("quikr_car.csv")

print("Data:")
data.display_rows()

print("\nTotal rows:",len(data.rows))

            

        
```
